```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
```

```python
# example of Simple Moving Average

N = 200

_, plts = plt.subplots(ncols=2, figsize=(16, 6))

for i, f in enumerate([np.sin, np.cos]):
    x = np.linspace(0, 1, N) * 2 * np.pi
    y = f(x)
    plts[i].plot(x, y, label="y")
    rand_y = y + (np.random.random(N) - 0.5) * 0.5
    plts[i].plot(x, rand_y, label="random + y")

    M = 10
    y = np.convolve(np.ones(M) / M, rand_y, mode="same")
    plts[i].plot(x, y, label="SMA")

    plts[i].legend()
```
