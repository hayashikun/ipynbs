```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
```

```python
# example of Gradation Color Plot

import matplotlib.cm as cm

_, plts = plt.subplots(ncols=3, nrows=3, figsize=(15, 15))

x = np.linspace(0, 1, 200) * 2 * np.pi

N = 20
y_list = [np.sin(x) * i for i in range(N)]

def random_color_map():
    return (np.random.rand(), np.random.rand(), np.random.rand(), 1)

for i, y in enumerate(y_list):
    plts[0][0].plot(x, y)
    plts[0][1].plot(x, y, color="red")
    plts[0][2].plot(x, y, color=["red", "blue", "green", "orange"][i % 4])
    plts[1][0].plot(x, y, color=cm.jet(i / N))
    plts[1][1].plot(x, y, color=cm.hsv(i / N))
    plts[1][2].plot(x, y, color=cm.ocean(i / N))
    plts[2][0].plot(x, y, color=random_color_map())
    plts[2][1].plot(x, y, color=random_color_map())
    plts[2][2].plot(x, y, color=random_color_map())
```
